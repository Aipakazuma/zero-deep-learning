```python
import time
```

```python
x, t = utils.get_data()
network = utils.init_network_mnist()
accuracy_cnt = 0
start = time.time()
for i in range(len(x)):
    y = utils.predict(network, x[i])
    p = np.argmax(y)
    if p == t[i]:
        accuracy_cnt += 1
end = start - time.time()
print('Accuracy: ', str(float(accuracy_cnt) / len(x)), ' time: ', end)
```

```python
x, t = utils.get_data()
network = utils.init_network_mnist()

batch_size = 100
accuracy_cnt = 0
start = time.time()
for i in range(0, len(x), batch_size):
    x_batch = x[i:i + batch_size]
    y_batch = utils.predict(network, x_batch)
    p = np.argmax(y_batch, axis=1)
    accuracy_cnt += np.sum(p == t[i:i + batch_size])
end = start - time.time()
print('Accuracy: ', str(float(accuracy_cnt) / len(x)), ' time: ', end)
```
